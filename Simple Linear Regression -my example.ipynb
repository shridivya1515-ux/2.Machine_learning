```python
import pandas as pd #Importing python library , pandas library is nothing but database
```

```python
dataset=pd.read_csv("Salary_Data.csv") #Reading the preprocessed data
```

```python
dataset # checking the data by using variables
```

```python
#splitting input & output values and storing it in variables
independent=dataset[["YearsExperience"]]
dependent=dataset[["Salary"]]
```

```python
independent #input variable
```

```python
dependent #output variable
```

```python
#splitting train set & test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Model Creation - Learning phase
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train) #substitution process to calculate weight and bias
```

```python
#to view weight(slope)
weight=regressor.coef_
weight
```

```python
#to view bias
bais=regressor.intercept_
bais
```

```python
#Assigning the variable to see the predicted value for original test data
y_pred=regressor.predict(x_test)
```

```python
#checking r2 score
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#save the deployment file
import pickle
filename ="final_linear_model.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))#deployment will be saved
```

```python
loaded_model=pickle.load(open("final_linear_model.sav",'rb'))#testing the model by loading
result=loaded_model.predict([[16]])
```
